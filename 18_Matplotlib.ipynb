# Die Matplotlib
Matplotlib ist eine Programmbibliothek für die Programmiersprache Python, die es erlaubt, mathematische Darstellungen aller Art anzufertigen. Die Bibliothek bietet mehrer intressante Funktionen. Im folgenden werden nur ein paar Funktionen betrachtet.
## Kurven-Plot

```python
import matplotlib.pyplot as plt # Einbinden der Bibliothek unter dem kürzel plt
ydate=[1,4,7,9,3,8,0,4]  # Liste der Y-Daten
xdate=[1,2,3,4,5,6,7,8]  # Liste der X-Daten (Anzahl der Elemente mus mit y übereinstimmen!)
plt.plot(xdate,ydate)   # Plot mit x und y Daten.
plt.show()              # Ausgabe des Plots
```

Optional können weitere Angaben wie z.b. Beschriftungen, Farbe und vieles mehr hinzugefügt werden.  
Mehr Infos unter [https://matplotlib.org/]("https://matplotlib.org/")  
Beispiel:

```python
ydate1=[1,4,7,9,3,8,0,4]  # Liste der Y-Daten
ydate2=[1,2,3,4,5,6,7,8]  # Liste der Y-Daten
ydate3=[2,2,2,4,4,5,5,9]  # Liste der Y-Daten
xdate=[1,2,3,4,5,6,7,8]  # Liste der X-Daten (Anzahl der Elemente mus mit y übereinstimmen!)
plt.figure(figsize=(10,6)) # Größe definieren
plt.plot(xdate,ydate1,linestyle='--',label="Linie 1")   # Wird gestrichelt dargestellt; Mit Angabe für Legende
plt.plot(xdate,ydate2,marker='o',label="Linie2")   # Punkte werden zusätzlich mit einem Punktdargestellt; Mit Angabe für Legende
plt.plot(xdate,ydate3,color="r",label="Linie3")   # Plot in der Farbe rot; Mit Angabe für Legende
plt.title("Matplotlib-Beispiel")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")
plt.legend()  # Legende soll dargestellt werden
plt.show()          
```

## Balkendiagramm
Eine weitere Funktion ist die Ausgabe eines Balkendiagramms.  
Beispiel:

```python
ydate1=[1,4,7,9,3,8,0,4]  # Liste der Y-Daten Linie 1
ydate2=[1,2,3,4,5,6,7,8]  # Liste der Y-Daten Linie 2
ydate3=[2,2,2,4,4,5,5,9]  # Liste der Y-Daten Linie 3
xdate=[1,2,3,4,5,6,7,8]  # Liste der X-Daten (Anzahl der Elemente mus mit y übereinstimmen!)
plt.figure(figsize=(15,8)) # Größe definieren
plt.bar(xdate,ydate1,label="Linie 1")   # Plot Linie 1 mit Label Angabe
plt.bar(xdate,ydate2,label="Linie2")   # Plot Linie 2
plt.bar(xdate,ydate3,color="r",label="Linie3")   # Plot Linie 3  mit Angabe der Farbe "rot"
plt.title("Matplotlib-Beispiel")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")
plt.legend()  # Legende soll dargestellt werden
plt.show()      
```

## Histgramme
Histogramme sind ein wichtiges Werkzeug für Statistik, da diese aufzeigen, wie häufig bestimmte Werte auftreten.  
Histogramme benötigen daher nur eine Liste als Parameter. Es empfielht sich allerdings, dass mit dem Parameter "bins" die Breite der Balken festgelegt wird.
Bei der Ausgabe des Plots wird der letzte Balken teilweise "verschoben" dargestellt.

```python
xdata=[1,1,1,1,1,1,2,2,3,4,5,6,3,5,6,3,4]
plt.hist(xdata,bins=60)
plt.show()
```

```python
import random
werte=[]
for i in range(10000):
    wuerfel1=random.randint(1,6)
    wuerfel2=random.randint(1,6)
    werte.append(wuerfel1+wuerfel2)
plt.hist(werte,30)
plt.show()
```
