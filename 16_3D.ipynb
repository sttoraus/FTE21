## Vorbreitung: Schneiden von Listen & Strings (Slicing)
Listen und Strings können geschnitten werden. Syntax:

```python
a[start:stop]  # items start through stop-1
a[start:]      # items start through the rest of the array
a[:stop]       # items from the beginning through stop-1
a[:]           # a copy of the whole array
```

```python
zeichenkette="123456789"
print(zeichenkette[2:6])  
print(zeichenkette[2:]) 
print(zeichenkette[2:6]) 
print(zeichenkette[:6])  
print(zeichenkette[:])
```

# Farbbilder
Farbbilder bestehen quasi aus drei Bildern, eins für jede Farbe (Rot, Grün, Blau).


```python
# Beispiel
import matplotlib.pyplot as plt
import numpy as np
matrix=np.zeros(shape=(3,3,3),dtype=np.uint8)  # Leere 3*3*3 Matrix erstellen
r=255
g=255
b=255
matrix[:,:,0]=r # füllt die erste 2D-Matrix mit dem Wert r
matrix[:,:,1]=g  # füllt die zweite 2D-Matrix mit dem Wert g
matrix[:,:,2]=b # füllt die dritte 2D-Matrix mit dem Wert b
plt.imshow(matrix)
plt.show()
```

```python
# Auf einzelne "Pixel" muss nun mit drei Indexen zugegriffen werden:
# Syntax  Liste[Zeile][Spalte][Ebene]
matrix[0][0][0]=0
plt.imshow(matrix)
plt.show()
```
